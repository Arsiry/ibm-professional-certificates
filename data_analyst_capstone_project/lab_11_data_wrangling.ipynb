### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
df.head()
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Display column data types
print("Column Data Types:")
print(df.dtypes)

# Count non-missing values for each column
print("\nNon-Missing Value Counts:")
print(df.count())

# Count missing values for each column
print("\nMissing Value Counts:")
print(df.isnull().sum())

# General summary of the dataset
print("\nDataset Info:")
print(df.info())
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Generate basic statistics for numerical columns
print("Basic Statistics for Numerical Columns:")
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Check unique values in the 'Country' column
print("Unique values in 'Country':")
print(df['Country'].unique())
```

```python
# Find rows with potential issues
print("Rows with potential inconsistencies in 'Country':")
print(df.loc[~df['Country'].str.match(r'^[A-Za-z ,]+$', na=False), 'Country'])
```

```python
# Missing values in 'Country'
missing_countries = df['Country'].isnull()
print("Rows with missing 'Country' values:")
print(df[missing_countries])
```

```python
# Irrelevant entries (e.g., numerical values or placeholders like 'Unknown')
irrelevant_countries = df['Country'].isin(['Unknown', 'N/A', 'None'])
print("Rows with irrelevant 'Country' values:")
print(df[irrelevant_countries])
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
# Define a mapping for the 'EdLevel' column
edlevel_mapping = {
    'Bachelor’S Degree (B.A., B.S., B.Eng., Etc.)': "Bachelor's Degree",
    'Master’S Degree (M.A., M.S., M.Eng., Mba, Etc.)': "Master's Degree",
    'Associate Degree (A.A., A.S., Etc.)': "Associate Degree",
    'Some College/University Study Without Earning A Degree': 'College/University Without A Degree',
    'Professional Degree (Jd, Md, Ph.D, Ed.D, Etc.)': "Professional Degree",
    'Secondary School (E.G. American High School, German Realschule Or Gymnasium, Etc.)': "Secondary School",
    'Primary/Elementary School': "Primary School",
    'Something Else': None,
    'nan': None
}

# Standardize the 'EdLevel' column
df['EdLevel'] = df['EdLevel'].str.strip().str.title()  # Normalize case and remove extra spaces
df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

# Display standardized values
print("Standardized 'EdLevel' column:")
print(df['EdLevel'].unique())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the EdLevel column using one-hot encoding.</h5>


```python
# One-hot encode the 'Employment' column
EdLevel_encoded = pd.get_dummies(df['EdLevel'], prefix='EdLevel')

# Add the encoded columns to the original DataFrame
df = pd.concat([df, EdLevel_encoded], axis=1)
df.head()
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
# Count the number of missing values in each column
missing_values = df.isnull().sum()

# Sort the columns by the number of missing values in descending order
sorted_missing = missing_values.sort_values(ascending=False)

# Display the columns with the highest number of missing values
print("Columns with the highest number of missing values:")
print(sorted_missing)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
# Using the mean for imputation
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean())

# Verify that there are no more missing values in the column
print("Missing values after imputation:")
print(df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
# Find the most frequent value (mode)
most_frequent_value = df['RemoteWork'].mode()[0]

# Fill missing values with the most frequent value
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent_value)

# Verify that there are no more missing values in the column
print("Missing values after imputation in 'RemoteWork':")
print(df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
from sklearn.preprocessing import MinMaxScaler

# Initialize Min-Max Scaler
scaler = MinMaxScaler()

# Reshape the data for scaling
df['ConvertedCompYearly_Scaled'] = scaler.fit_transform(df[['ConvertedCompYearly']])
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
# Apply log transformation
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'])
```

```python
df.head()
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
print(df['YearsCodePro'].unique())
```

```python
# Define a mapping for the 'EdLevel' column
YearsCodePro_mapping = {
    'Less than 1 year': 0,
    'More than 50 years': 51
}

# Standardize the 'EdLevel' column
df['EdLYearsCodeProevel'] = df['YearsCodePro'].str.strip().str.title()  # Normalize case and remove extra spaces
df['YearsCodePro'] = df['YearsCodePro'].replace(YearsCodePro_mapping)

# Display standardized values
print("Standardized 'YearsCodePro' column:")
print(df['YearsCodePro'].unique())
```

```python
# Преобразование в числовой тип, с обработкой NaN
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
```

```python
# Вычисление медианы (игнорируя NaN)
median_value = df['YearsCodePro'].median()

# Замена NaN на медиану
df['YearsCodePro'].fillna(median_value, inplace=True)
```

```python
# Define bins and labels for experience levels
bins = [-1, 0, 2, 5, 10, float('inf')]  # Define bins for experience ranges
labels = ['Beginner', 'Novice', 'Intermediate', 'Advanced', 'Expert']  # Corresponding labels

# Create the ExperienceLevel column
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels)
```

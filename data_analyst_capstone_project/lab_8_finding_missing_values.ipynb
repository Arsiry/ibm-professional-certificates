## Tasks


<h2>1. Load the Dataset</h2>
<p>
We use the <code>pandas.read_csv()</code> function for reading CSV files. However, in this version of the lab, which operates on JupyterLite, the dataset needs to be downloaded to the interface using the provided code below.
</p>


The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
df.head()
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
# Display basic information about the dataset
print("Basic Information about the Dataset:")
df.info()
```

```python
# Display summary statistics of the dataset
print("\nSummary Statistics of the Dataset:")
df.describe()
```

```python
# Display information about categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
if len(categorical_columns) > 0:
    print("\nValue Counts for Categorical Columns:")
    for column in categorical_columns:
        print(f"\n{column}:\n{df[column].value_counts()}")
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
# Check for missing values in all columns
missing_values = df.isnull().sum()

# Display columns with missing values and their counts
print("Missing Values in Each Column:")
print(missing_values[missing_values > 0])
```

```python
# Display the percentage of missing values for each column
missing_percentage = (missing_values / len(df)) * 100
print("\nPercentage of Missing Values in Each Column:")
print(missing_percentage[missing_percentage > 0])
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(12, 8))

# Plot a heatmap to visualize missing values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

# Add a title to the plot
plt.title("Missing Values Heatmap", fontsize=16)

# Show the plot
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `EdLevel`).


```python
# Count the number of missing rows for the column 'Employment'
missing_EdLevel = df['EdLevel'].isnull().sum()

# Display the result
print(f"Number of missing rows in the 'EdLevel' column: {missing_EdLevel}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `EdLevel`).


```python
# Find the most frequent value in the 'Employment' column
most_frequent_value = df['EdLevel'].mode()[0]

# Display the result
print(f"The most frequent value in the 'EdLevel' column is: {most_frequent_value}")
```

##### Task 6: Impute missing values in the `EdLevel` column with the most frequent value.



```python
# Identify the most frequent value in the 'EdLevel' column
most_frequent_value = df['EdLevel'].mode()[0]

# Impute missing values in the 'EdLevel' column with the most frequent value
df['EdLevel'].fillna(most_frequent_value, inplace=True)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `EdLevel`).


```python
## Write your # Set the figure size
plt.figure(figsize=(10, 6))

# Plot the distribution of the 'EdLevel' column
sns.countplot(data=df, x='EdLevel', order=df['EdLevel'].value_counts().index)

# Add labels and a title
plt.title("Distribution of 'EdLevel' Column After Imputation", fontsize=16)
plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Count", fontsize=14)

# Rotate x-axis labels for better readability (if necessary)
plt.xticks(rotation=45)

# Show the plot
plt.show() 
```

```python
def process_complex_column(df, column_name):
    """
    Process the  column to count occurrences and create binary columns for each unique value.

    Args:
        df (pd.DataFrame): The input dataframe.
        column_name (str): The name of the column to process.

    Returns:
        pd.DataFrame, pd.Series: Updated dataframe and counts of unique value.
    """
    # Split and count unique endorsements
    value_counts = df[column_name].str.split(';').explode().value_counts()

    # Generate binary columns for each unique endorsement
    unique_values = value_counts.index
    for col in unique_values:
        binary_column_name = f"{column_name}_{col.replace(' ', '').replace(':', '').replace(';', '').replace(',', '')}"
        df[binary_column_name] = df[column_name].str.contains(col, case=False, na=False).astype(int)

    df = df.drop(column_name, axis=1)

    return df, value_counts
```

```python
# Missing values

# Define the critical columns
critical_columns = ['YearsCode', 'YearsCodePro', 'CompTotal','JobSat', 'WorkExp',
                    'JobSatPoints_1', 'JobSatPoints_4','JobSatPoints_5','JobSatPoints_6','JobSatPoints_7',
                    'JobSatPoints_8','JobSatPoints_9','JobSatPoints_10','JobSatPoints_11',]

# Check the percentage of missing values in critical columns
missing_percentage = df_about_respondents[critical_columns].isnull().mean() * 100
print("Percentage of missing values in critical columns:")
print(missing_percentage)
```

```python
# Handling missing values

for column in critical_columns:    
    df_about_respondents[column] = df_about_respondents[column].fillna(0)
```

```python
# Define experience ranges for YearsCodePro
def categorize_experience(years):
    try:
        years = float(years)
        if years <= 5:
            return '0-5'
        elif years <= 10:
            return '5-10'
        elif years <= 20:
            return '10-20'
        else:
            return '>20'
    except ValueError:
        return None

df_about_respondents['ExperienceRange'] = df_about_respondents['YearsCodePro'].apply(categorize_experience)
df_about_respondents['ExperienceRange'].unique()
```
